```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Read in data
df = pd.read_csv("data\wingspan birds.csv")

#Filter rows and columns
df = df[df["Common name"].notnull()].iloc[:, 0:26]


print(df.head(10).to_string())
```

### Cleaning data

```python
#Find nulls in 'Total food cost' column
print(df[df["Total food cost"].isnull()].to_string())
```

```python
#Replace nulls in food cost columns
df.iloc[:, 16:23] = df.iloc[:, 16:23].fillna(0)
df
```

```python
df["Total food cost"] = np.where(df["/ (food cost)"] == 'X', 1, df.iloc[:, 16:23].sum(axis=1))
df.head(50)
```
